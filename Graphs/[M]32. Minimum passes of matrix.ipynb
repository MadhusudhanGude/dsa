```python
# O(w*h)T and O(w*h)S

def min_passes(matrix):
    passes = convert_negatives(matrix)
    is_negative = cointains_negatives(matrix)
    return passes - 1 if not is_negative else -1

def convert_negatives(matrix):
    curr_positives = []

    for row in range(len(matrix)):
        for col in range(len(row)):
            if matrix[row][col] > 0:
                curr_positives.append((row, col))
    
    passes = 0

    while len(curr_positives) > 0:
        pos_count = len(curr_positives)

        while pos_count > 0:
            curr_row, curr_col = curr_positives.pop(0)

            neighbours_of_curr = get_neighbours(curr_row, curr_col, matrix)
            for each_neighbour in neighbours_of_curr:
                n_row, n_col = each_neighbour
                if matrix[n_row][n_col] < 0:
                    matrix[n_row][n_col] *= -1
                    curr_positives.append([n_row, n_col])

            pos_count -=1
        passes += 1
    return passes
    
def get_neighbours(row, col, matrix):
    neighbours = []
    if row > 0:
        neighbours.append(row-1, col)
    if row < len(matrix)-1:
        neighbours.append(row+1, col)
    if col > 0:
        neighbours.append(row, col-1)
    if col < len(matrix[0]):
        neighbours.append(row, col+1)
    return neighbours

def cointains_negatives(matrix):
    for row  in matrix:
        for value in row:
            if value < 0:
                return True
    return False
```
