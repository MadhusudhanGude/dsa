```python
class Node:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add_child(self, value):
        self.children.append(Node(value))

    # O(V+E)T and O(v)S
    def bfs(self, array):
        stack = [self]
        while len(stack) > 0:
            current = stack.pop(0)
            array.append(current.name)
            for child in current.children():
                stack.append(child)
            
        return array
```
