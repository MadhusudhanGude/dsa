```python
# O(wh)T and O(wh)S
def river_sizes(matrix):
    sizes = []
    visited_matrix = [[False for item in row] for row in matrix]
    for i in range(len(matrix)):
        for j in range(len(matrix)):
            traverse(i, j, matrix, visited_matrix, sizes)
    return sizes

def traverse(i, j, matrix, visited_matrix, sizes):
    current_size = 0
    nodes_to_explore = [ [[i][j]] ]
    
    while len(nodes_to_explore) > 0:
        current_node = nodes_to_explore.pop()
        i, j = current_node[0], current_node[1]

        if visited_matrix[i][j]:
            # When it is already visited
            continue
            
        visited_matrix[i][j] = True

        if matrix[i][j] == 0:
            continue
        
        current_size += 1

        unvisited_neighbours = find_unvisited_neighbours(i, j, matrix, visited)
        for each_unvisited_neighbour in unvisited_neighbours:
            nodes_to_explore.append(each_unvisited_neighbour)

    if current_size > 0:
        sizes.append(current_size)


def find_unvisited_neighbours(i, j, matrix, visited):
    unvisited_neighbours = []
    if i > 0 and not visited[i-1][j]:
        unvisited_neighbours.append([[i-1][j]])
    
    if i < (len(matrix)-1) and not visited[i+1][j]:
        unvisited_neighbours.append([[i+1][j]])
    
    if j > 0 and not visited[i][j-1]:
        unvisited_neighbours.append([[i][j-1]])

    if j < len(matrix[0]) - 1 and not visited[i][j+1]:
        unvisited_neighbours.append([[i][j+1]])
    return unvisited_neighbours
```
