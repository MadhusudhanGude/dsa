```python
# O(wh)Time
# O(wh)S(for cal trace)+O(wh)S(for auxilry matrix) ~= O(wh)S
def remove_islands(matrix):
    ones_connected_to_border = [[False for col in row] for row in matrix]
    for row in matrix:
        for col in row:
            is_row_border = row == 0 or row == len(matrix)-1
            is_col_border = col == 0 or col ==  len(row)-1
            is_border = is_row_border or is_col_border
            
            if is_border and matrix[row][col]==1:
                find_connected_to_border(matrix, row, col, ones_connected_to_border)
            else:
                continue

    for row in range(1, len(matrix)-1):
        for col in range(1, len(row)-1):
            if ones_connected_to_border[row][col]:
                continue
            matrix[row][col] = 0
    return matrix
    
def find_connected_to_border(matrix, row, col, ones_connected_to_border):
    stack = [(row, col)]

    while len(stack) > 1:
        cur_row, cur_col = stack.pop()
        already_visited = ones_connected_to_border[cur_row][cur_col]
        if already_visited:
            continue
        ones_connected_to_border[cur_row][cur_col] = True

        neighbours = get_neighbours(matrix, cur_row, cur_col)

        for each_neighbour in neighbours:
            n_row, n_col =  each_neighbour
            if matrix[n_row][n_col] != 1:
                continue

            stack.append(each_neighbour)
        
    
def get_neighbours(matrix, row, col):
    neighbours = []
    total_rows  = len(matrix)
    total_cols  = len(matrix[0])
    # UP neighbour
    if row -1 >= 0:
        neighbours.append(row -1 , col)
    # DOWN neighbour
    if row + 1 < total_rows:
        neighbours.append(row+1 , col)

    # LEFT neighbour
    if col -1 >= 0:
        neighbours.append(row, col-1)
    # RIGHT neighbour
    if col + 1 < total_cols:
        neighbours.append(row , col+1)

    return neighbours
        

```
