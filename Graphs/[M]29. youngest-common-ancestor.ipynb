```python
# O(d)T and O(1)S
def get_youngest_ancestor(top_ancestor, desc_one, desc_two):
    desc_one_depth = get_depth_of_asc(desc_one, top_ancestor)
    desc_two_depth = get_depth_of_asc(desc_two, top_ancestor)

    if desc_one_depth > desc_two_depth:
        traceback_ancestor(desc_one, desc_two, desc_one_depth-desc_two_depth)
    else:
        traceback_ancestor(desc_two, desc_one, desc_two_depth-desc_one_depth)


def get_depth_of_asc(descendant, top_ansc):
    current_depth = 0
    while descendant != top_ansc:
        current_depth += 1
        descendant = descendant.ancestor
    return current_depth

def traceback_ancestor(lower_desc, higher_desc, diff):
    while diff != 0:
        lower_desc = lower_desc.ansector
        diff -= 1
    while lower_desc != higher_desc:
        lower_desc = lower_desc.ansector
        higher_desc = higher_desc.ansector
    return lower_desc
```
