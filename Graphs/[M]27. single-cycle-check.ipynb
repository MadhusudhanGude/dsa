```python
# O(n)T and O(1)S
def has_single_cycle(array):
    current_index = 0 
    num_of_ele_visited = 0
    while num_of_ele_visited < len(array):
        if num_of_ele_visited > 0 and current_index == 0:
            return False
        num_of_ele_visited += 1
        current_index = get_next_idx(current_index, array)
    return current_index == 0

def get_next_idx(current_index, array):
    jump = array[current_index]
    next_idx = (current_index + jump) % len(array)
    return next_idx if next_idx >= 0 else next_idx + len(array)


has_single_cycle([2,3,1,-4,-4,2])
```
