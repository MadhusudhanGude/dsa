```python
# O(V+E)T
# O(N)S
class Node:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add_child(value):
        self.children.append(Node(value))

    # O(V+E)T and O(v)S
    def dfs(self, array):
        array.append(self.name)
        for child in self.children():
            child.dfs(array)
        
        return array
```
