```python
# https://www.youtube.com/watch?v=FvnTWDKT_ck

# O(1)S and O(n log(n))T
def h_index(citations):
    # O(n log(n))T
    citations.sort()

    total_len = len(citations)

    # O(n)T
    for idx, val in enumerate(citations):
        if val >= total_len - idx:
            return total_len - idx

    return -1


citations = [3,0,6,1,5]
print(h_index(citations))
# Output: 3
```

```python
# O(n)S and O(n)T
def h_index(citations):
    total_len = len(citations)
    total_occurences = [0] * (total_len + 1)

    for val in citations:
        if val > total_len:
            total_occurences[total_len] += 1
        else:
            total_occurences[val] += 1

    total_sum = 0
    for idx in reversed(range(0, total_len + 1)):
        total_sum += total_occurences[idx]
        if total_sum >= idx:
            return idx

    return -1


citations = [3,0,6,1,5]
print(h_index(citations))
# Output: 3
```
