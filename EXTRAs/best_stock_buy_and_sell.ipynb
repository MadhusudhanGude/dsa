```python
# O(N)T and O(1)S
def max_profit(prices):
    # Buy is always on the left side and sell is always on the right side
    buy_idx, sell_idx = 0, 1

    max_profit = 0
    while sell_idx < len(prices):
        # maximise profits by selling for higher profit
        if  prices[sell_idx] > prices[buy_idx]:
            profit =  prices[sell_idx] - prices[buy_idx]
            max_profit = max(max_profit, profit)
        else:
            buy_idx = sell_idx

        sell_idx += 1

    return max_profit


max_profit([7,1,5,3,6,4])
```
