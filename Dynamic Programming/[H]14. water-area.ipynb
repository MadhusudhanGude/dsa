```python
# O(N)T and O(N)S

def water_area(heights):
    maxes = [0 for height in heights]
    print(maxes)
    left_max = 0
    for i in range(len(heights)):
        height = heights[i]
        maxes[i] = left_max
        left_max = max(left_max, height)
    print(maxes)
    right_max = 0
    for i in reversed(range(len(heights))):
        height = heights[i]
        min_height = min(right_max, maxes[i])
        if height < min_height:
            maxes[i] = min_height - height
        else:
            maxes[i] = 0
        right_max = max(right_max, height)
    print(maxes)
        
    return sum(maxes)

water_area([0,8,0,0,5,0,0,10,0,0,1,1,0,3])
```
