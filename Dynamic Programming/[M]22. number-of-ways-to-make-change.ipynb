```python

# O(n)S and O(nd)T
def make_change(target, demons):
    ways = [0 for i in range(target+1)]
    ways[0] = 1

    for denom in demons:
        for amount in range(1, target+1):
            if denom <= amount:
                ways[amount] += ways[amount - denom]
                
    return ways[-1]

make_change(10, [1,5,10,25])
```
