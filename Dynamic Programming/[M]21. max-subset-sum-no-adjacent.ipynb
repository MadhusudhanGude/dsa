```python
# O(n)T and O(n)S
def max_subset_sum_no_adjacent(array):
    if len(array)==0:
        return -1

    elif len(array)==1:
        return array[0]

    max_subset = array[:]
    max_subset[1] = max(array[0], array[1])

    for i in range(2, len(array)):
        max_subset[i] =  max(max_subset[i-1], max_subset[i-2]+max_subset[i])
    return max_subset[-1]

max_subset_sum_no_adjacent([7,10,12,7,9,14])
```

```python
# O(n)T and O(1)S
def max_subset_sum_no_adjacent(array):
    if len(array)==0:
        return -1
        
    elif len(array)==1:
        return array[0]

    second = array[0]
    first = max(array[0], array[1])

    for i in range(2, len(array)):
        current = max(first, second + array[i])
        second = first
        first = current

    return current

max_subset_sum_no_adjacent([7,10,12,7,9,14])
```
