```python
# O(2^(n+m)) T and O(n+m) S
def num_ways_to_traverse_graph(width, height):
    if width == 1 or height == 1:
        return 1

    return num_ways_to_traverse_graph(width -1 , height) + num_ways_to_traverse_graph(width, height - 1)
num_ways_to_traverse_graph(3, 4)
```

```python
# O(n * m) T and O(n * m) S
def num_ways_to_traverse_graph(width, height):
    num_of_ways = [[0 for i in range(width+1)] for i in range(height+1)]
    for width_idx in range(1, width + 1):
        for height_idx in range(1, height + 1):
            if height_idx == 1 or height_idx == 1:
                num_of_ways[height_idx][width_idx] = 1
            else:
                ways_up = num_of_ways[height_idx - 1][width_idx]
                ways_left = num_of_ways[height_idx][width_idx - 1]
                num_of_ways[height_idx][width_idx] = ways_up + ways_left
    return num_of_ways[height][width]
    
num_ways_to_traverse_graph(3, 4)
```

```python
# O(n + m) T and O(1) S
def num_ways_to_traverse_graph(width, height):
    x_distance_to_corner = width - 1
    y_distance_to_corner = height - 1

    numerator = factorial(x_distance_to_corner + y_distance_to_corner)
    denominator = factorial(x_distance_to_corner) * factorial(y_distance_to_corner)

    return numerator // denominator

def factorial(num):
    result = 1
    for n in range(2, num+1):
        result *= n
    return result
    
num_ways_to_traverse_graph(3, 4)
```
