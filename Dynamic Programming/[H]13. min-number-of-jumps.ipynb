```python
# O(N^2)T and O(N)S
def min_num_of_jumps(array):
    jumps = [float('inf') for x in array]
    jumps[0] = 0
    for i in range(1, len(array)):
        for j in range(0, i):
            if array[j] + j >= i:
                jumps[i] = min(jumps[j] + 1, jumps[i])
    print(jumps)
    return jumps[-1]
min_num_of_jumps([3,4,2,1,2,3,7,1,1,1,3])
```

```python
# O(1)S and O(N)T
def min_num_of_jumps(array):
    if len(array) == 1:
        return 0

    jumps = 0
    max_reach = array[0]
    steps = array[0]
    for i in range(1, len(array) - 1):
        max_reach =  max(max_reach,  i + array[i])
        steps -= 1
        if steps == 0 :
            jumps += 1
            steps = max_reach - i
    return jumps + 1
min_num_of_jumps([3,4,2,1,2,3,7,1,1,1,3])   
```
