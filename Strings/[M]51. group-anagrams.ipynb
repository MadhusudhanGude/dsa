```python
# O(w * n log(n))(line 4 sorting) + O(w log(w)(w-indices) * n)  = O(w * n log(n)) + O(w log(w) * n)T
# O(wn)S
def group_anagrams(words):
    # print(sorted(words))
    sorted_words = [''.join(sorted(w)) for w in words]
    indices = [i for i in range(len(words))]
    # print(sorted_words)
    # print(indices)
    indices.sort(key= lambda x: sorted_words[x])
    # print(sorted_words)
    # print(indices)

    result = []
    current_anagram = sorted_words[indices[0]]
    current_anagram_list = []

    for i in indices:
        word = words[i]
        sorted_word = sorted_words[i]
        if sorted_word == current_anagram:
            current_anagram_list.append(word)
        else:
            result.append(current_anagram_list)
            current_anagram = sorted_word
            current_anagram_list = [word]

    result.append(current_anagram_list)
    return result
group_anagrams(['yo','oy', 'act', 'flop', 'tac', 'cat', 'plof'])
```

```python
# O(w * n log(n))T
# O(WN)S
def group_anagrams(words):
    anagrams = {}
    for word in words:
        sorted_word = ''.join(sorted(word))
        if sorted_word in anagrams:
            anagrams[sorted_word].append(word)
        else:
            anagrams[sorted_word] = [word]
    return list(anagrams.values())
group_anagrams(['yo','oy', 'act', 'flop', 'tac', 'cat', 'plof'])
```
