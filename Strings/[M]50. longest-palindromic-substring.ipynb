```python
# O(N^2)T and O(1)S
def get_longest_pstr(string):
    # [start_idx, last_idx]
    current_longest_idx = [0,1]
    for i in range(1, len(string)):
        odd_str_idx = get_palindrome(i-1, i+1, string)
        even_str_idx = get_palindrome(i-1, i, string)
        str_idx = max(odd_str_idx, even_str_idx, key=lambda x:x[1]-x[0])
        current_longest_idx = max(current_longest_idx, str_idx, key=lambda x:x[1]-x[0])

    return string[current_longest_idx[0]:current_longest_idx[1]]

def get_palindrome(left_idx, right_idx, string):
    while left_idx > 0 and right_idx < len(string):
        if not string[left_idx] == string[right_idx]:
            break
        left_idx -= 1
        right_idx += 1

    return [left_idx+1, right_idx]

get_longest_pstr('abaxyzzyxf')
```
