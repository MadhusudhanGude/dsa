```python
def valid_ip_addresses(string):
    result = []

    for i in range(1, min(len(string)-1, 4) ):
        new_ip_comb = ['', '', '', '']

        new_ip_comb[0] = string[:i]
        if not is_valid_ip_part(new_ip_comb[0]):
            continue
        
        for j in range(i+1, i + min(len(string)-i, 4)):
            new_ip_comb[1] = string[i:j]
            if not is_valid_ip_part(new_ip_comb[1]):
                continue

            for k in range(j+1, j + min(len(string)-j, 4)):
                new_ip_comb[2] = string[j:k]
                new_ip_comb[3] = string[k:]
                if not (is_valid_ip_part(new_ip_comb[2]) and is_valid_ip_part(new_ip_comb[3])):
                    continue
        
                result.append('.'.join(new_ip_comb))

    return result

def is_valid_ip_part(string):
    if int(string) > 255:
        return False
    return len(string) == len(str(int(string)))

valid_ip_addresses('1921680')
```
