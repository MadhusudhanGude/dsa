```python
# O(N)S and O(N)T
def caesar_cipher_encryptor(string, key):
    new_string = []
    alphabets = list('abcdefghijklmnopqrstuvwxyz')
    key = key % 26
    for letter in string:
        new_string.append(get_new_letter(letter, key, alphabets))
    return ''.join(new_string)

def get_new_letter(letter, key, alphabets):
    new_letter = alphabets.index(letter) + key
    return alphabets[new_letter] if new_letter <= 25 else alphabets[-1 + new_letter%25]


caesar_cipher_encryptor('xyz', 3)
```
