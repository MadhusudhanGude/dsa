```python
# O(n) + O(n) ~= O(n)T and O(1)S
def get_first_non_repeating_char(string):
    characters = {}
    for char in string:
        characters[char] = characters.get(char, 0) + 1

    for i in range(len(string)):
        if characters[string[i]] == 1:
            return i, string[i]

get_first_non_repeating_char('shdlusausjkfjdjkdjkjkjk')
```
