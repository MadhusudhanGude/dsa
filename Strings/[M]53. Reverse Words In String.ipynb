```python
# O(n)S and O(n)T
def reverse_words(string):

    new_word_start_idx = 0
    words_list = []
    for index in range(len(string)):

        if string[index] == ' ':
            words_list.append(string[new_word_start_idx:index])
            new_word_start_idx = index

        elif string[new_word_start_idx] == ' ':
            words_list.append(' ')
            new_word_start_idx = index
        
    words_list.append(string[new_word_start_idx:])
    words = reverse_list(words_list)
    return ''.join(words)

def reverse_list(words_list):
    left = 0
    right = len(words_list)-1
    
    while left < right:
        words_list[left], words_list[right] = words_list[right], words_list[left]
        left += 1
        right -= 1

    return words_list
reverse_words('Alogexpert is  the best!')
```
