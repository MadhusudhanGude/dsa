```python
# O(N**2)T and O(N)S
def is_palindrome(string):
    reversed_str = ''
    for i in reversed(range(len(string))):
        reversed_str += string[i]
    return reversed_str == string

is_palindrome('abcba')
```

```python
# O(N)T and O(N)S
def is_palindrome(string):
    reversed_str = []
    for i in reversed(range(len(string))):
        reversed_str.append(string[i])
    return string == ''.join(reversed_str)

is_palindrome('abcba')
```

```python
# Using recursion
# O(N)T and O(N)S
def is_palindrome(string, i=0):
    j = len(string)-1 - i
    return True if i >= j else string[i] == string[j]  and is_palindrome(string, i+1)

is_palindrome('abcba')
```

```python
# O(N)T and O(1)S
def is_palindrome(string):
    l_ptr = 0
    r_ptr = len(string)-1
    while l_ptr <= r_ptr:
        if not string[l_ptr] == string[r_ptr]:
            return False
        l_ptr += 1
        r_ptr -= 1
    return True

is_palindrome('abcba')
```
