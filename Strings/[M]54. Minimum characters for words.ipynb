```python
# O(n*L)T and O(c)S
def find_min_char_for_words(words):
    max_char_freq_count = {}
    result = []
    for word in words:
        char_freq = get_char_freq(word)
        update_char_freq_count(char_freq, max_char_freq_count)

    for char in max_char_freq_count:
        for _ in range(max_char_freq_count[char]):
            result.append(char)

    return result
    
def update_char_freq_count(char_freq, max_char_freq_count):
    for char in char_freq:
        freq = char_freq[char]

        if char in max_char_freq_count:
            max_char_freq_count[char] = max(freq, max_char_freq_count[char])
        else:
            max_char_freq_count[char] = freq

def get_char_freq(string):
    char_freq = {}

    for char in string:
        if not char in char_freq:
            char_freq[char] = 0
        char_freq[char] += 1
    return char_freq

find_min_char_for_words(["this", "that", "did", "deed"])
```
