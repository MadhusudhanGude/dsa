```python
# O(n+m)T and O(n)S
def underscorify_substring(string, sub_string):
    locations = get_locations(string, sub_string)
    collapsed_locations = collapse_locations(locations)
    return under_scorfiy(string, collapsed_locations)

def under_scorfiy(string, locations):
    locations_idx = 0
    string_idx = 0
    in_between_underscores = False
    final_chars = []
    i=0
    while string_idx < len(string) and locations_idx < len(locations):
        if string_idx == locations[locations_idx][i]:
            final_chars.append('_')
            i = 0 if i==1 else 1
            in_between_underscores = not in_between_underscores
            if not in_between_underscores:
                locations_idx += 1
        final_chars.append(string[string_idx])
        string_idx += 1
    
    if string_idx < len(string):
        final_chars.append(string[string_idx:])
    elif locations_idx < len(locations):
        final_chars.append('_')

    return ''.join(final_chars)


def collapse_locations(locations):
    if not locations:
        return locations
    collapsed_locations = [locations[0]]
    previous = locations[0]

    for i in range(1, len(locations)):
        current = locations[i]
        if current[0] <= previous[1]:
            previous[1] = current[1]
        else:
            collapsed_locations.append(current)
            previous = current

    return collapsed_locations

def get_locations(string, sub_string):
    locations = []
    idx = 0
    while idx < len(string):
        start_idx = string.find(sub_string, idx)
        if start_idx != -1:
            locations.append([start_idx, start_idx+len(sub_string)])
            idx = start_idx+1
        else:
            break
    return locations

string = 'testthis is a testtest to see if testtesttest it works'
sub_string = 'test'
underscorify_substring(string, sub_string)
```
