```python
# O(c*(n+m))T and O(c)S
def generate_document(characters, document):
    characters_count = {}

    for character in document:
        if character in characters_count:
            continue

        char_freq = get_char_count(character, characters)
        doc_freq = get_char_count(character, document)

        if char_freq < doc_freq:
            return False

        characters_count[character] = True
    return True

def get_char_count(char, target):
    freq = 0
    for each_char in target:
        if char == each_char:
            freq += 1
    return freq
```

```python
# O(n+m)T and O(c)S
def generate_document(characters, document):
    characters_count = {}

    for character in characters:
        if characters not in characters_count:
            characters_count[characters] = 0
        characters_count[characters] += 1

    for character in document:
        if character not in characters_count or characters_count[character] == 0:
            return False
        characters_count[characters] -= 1

    return True

```
