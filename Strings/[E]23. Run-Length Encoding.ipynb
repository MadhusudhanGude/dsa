```python
# O(n)for storing the result; can we O(2n) in worst case if the characters are non-repeating ~= O(n)S
# O(n) for first for loop +O(n) for joining the result ~= O(n)T
def running_length_encoding(string):
    result = []
    current_running_length = 1
    for i in range(1, len(string)):
        current_chr = string[i]
        prev_chr = string[i-1]

        if current_chr != prev_chr or current_running_length == 9:
            result.append(str(current_running_length))
            result.append(prev_chr)
            current_running_length = 0
        
        current_running_length += 1
    
    result.append(str(current_running_length))
    result.append(current_chr)

    return ''.join(result)
    
running_length_encoding('AAAAAAAAAAABBUUUUUY')
```
