```python
# O(N)T and O(N)S
def array_of_products(array):
    result = [1] * len(array)
    print(result)
    left_array_product = 1
    for idx in range(len(array)):
        print(idx)
        result[idx] *= left_array_product
        left_array_product *= array[idx]
    print(result)
    

    right_array_product = 1
    for idx in reversed(range(len(array))):
        result[idx] *= right_array_product
        right_array_product *= array[idx]
    print(left_array_product)
    print(right_array_product)
    return result
    
array_of_products([5,1,4,2])
```
