```python
# O(N)T & O(N)S
def find_duplicate(array):
    seen = set()
    for ele in array:
        if ele in seen:
            return ele
        seen.add(ele)
    return -1

find_duplicate([2,1,5,3,3,2,4])
```

```python
# O(N)T & O(1)S
def find_duplicate(array):

    for value in array:
        if array[abs(value)-1] < 0:
            return abs(value)
        array[value-1] *= -1

    return -1

find_duplicate([2,1,5,3,3,2,4])
```
