```python
# O(N)T and O(N)S
def min_rewards(array):
    rewards = [1 for _ in array]
    local_mins = get_local_mins(array)
    print(local_mins)
    for local_min in local_mins:
        expend_from_localmin(local_min, array, rewards)
        print(rewards)
    return sum(rewards)

def get_local_mins(array):
    if len(array)==1:
        return [0]
    local_mins = []

    for i in range(len(array)):
        if i == 0 and array[i] < array[i+1]:
            local_mins.append(i)
        if i == len(array)-1 and array[i] < array[i-1]:
            local_mins.append(i)
        if i ==0 or i == len(array)-1:
            continue
        if array[i] < array[i-1] and array[i] < array[i+1]:
            local_mins.append(i)
    return local_mins

def expend_from_localmin(local_min, array, rewards):
    left_idx = local_min -1
    while left_idx >= 0 and array[left_idx] > array[left_idx+1]:
        rewards[left_idx] = max(rewards[left_idx], rewards[left_idx+1]+1)
        left_idx -= 1
    right_idx = local_min + 1
    while right_idx < len(array) and array[right_idx] > array[right_idx-1]:
        rewards[right_idx] = rewards[right_idx-1]+1
        right_idx += 1

min_rewards([8,4,2,1,3,6,7,9,5, 4])
```
