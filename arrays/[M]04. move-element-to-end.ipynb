```python
def move_ele_to_end(array, ele_to_move):
    i = 0
    j = len(array)-1

    while i < j:
        while i < j and array[j] == ele_to_move:
            j -= 1

        if array[i] == ele_to_move:
            array[i], array[j] = array[j], array[i]

        i += 1
    
    return array

move_ele_to_end([2,1,2,2,2,3,4,2], 2)
```
