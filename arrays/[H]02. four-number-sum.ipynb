```python
# O(n^2)T and O(n^2)S
def four_num_sum(array, sum):
    result = []
    qudaraple_sum = {}
    for i in range(1, len(array)-1):
        for j in range(i+1, len(array)):
            two_sum = array[j]+array[i]
            remaining_sum = sum - two_sum
            if remaining_sum in qudaraple_sum:
                # import pdb; pdb.set_trace()
                for each_pair in qudaraple_sum[remaining_sum]:
                    quad_pair = [array[j],array[i]]
                    # print(each_pair)
                    quad_pair += each_pair
                    result.append(quad_pair)

        for k in range(0, i):
            key = array[k]+array[i]
            if key in qudaraple_sum:
                qudaraple_sum[key].append([array[k],array[i]])
            else:
                qudaraple_sum[key] = [[array[k],array[i]]]

    return result
four_num_sum([7,6,4,-1,1,2], 16)
```
