```python
# O(N)T and O(1)S
def is_monotonic(array):

    is_increasing_array = True
    is_decreasing_array = True

    for index in range(2,len(array)):
        if array[index] > array[index-1]:
            is_decreasing_array = False
        if array[index] < array[index-1]:
            is_increasing_array = False

    return is_increasing_array or is_decreasing_array

is_monotonic([-1,-5, -17, -102])
        
```
