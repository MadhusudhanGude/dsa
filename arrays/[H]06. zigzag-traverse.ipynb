```python
def zigzag_traverse(array):
    result = []
    width = len(array[0])-1
    height = len(array)-1
    going_down = True
    row, col = 0, 0
    while not is_outof_bounds(row, col, height, width):
        result.append(array[row][col])
        if going_down:
            if col == 0 or row == height:
                going_down = False
                if col ==  0:
                    row += 1
                else:
                    col += 1
            else:
                row += 1
                col -= 1
        else:
            if row == 0 or col == width:
                going_down = True
                if col == width:
                    row += 1
                else:
                    col += 1

            else:
                row -= 1
                col += 1
    return result

def is_outof_bounds(row, col, height, width):
    return row < 0 or row > height or col < 0 or col > width

zigzag_traverse([[1,3,4,10],
                 [2,5,9,11],
                 [6,8,12,15],
                 [7,13,14,16]])
```
