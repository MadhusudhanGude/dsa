```python
def largest_range(array):
    nums = {}
    best_range = []
    longest_length = 0

    for num in array:
        nums[num] = True
    
    for cur_num in array:

        if not nums[cur_num]:
            continue

        nums[cur_num] = False
        cureent_length = 1  
        left = cur_num - 1
        right = cur_num + 1

        while left in nums:
            nums[left]=False
            left -= 1
            cureent_length += 1
        
        while right in nums:
            nums[right]=False
            right += 1
            cureent_length += 1

        if cureent_length > longest_length:
            longest_length = cureent_length
            best_range = [left+1, right-1]
        
    return best_range

largest_range([1,11,3,0,15,5,2,4,10,7,12,6])
```
