```python
# O(N)T and O(N)S : N is number of elements
def spiral_traverse(array):
    
    result = []
    start_col = 0
    end_col = len(array[0])-1
    start_row = 0
    end_row = len(array)-1

    while start_col <= end_col and start_row <= end_row:
        for col in range(start_col, end_col+1):
            result.append(array[start_row][col])
        
        for row in range(start_row+1, end_row+1):
            result.append(array[row][end_col])

        for col in reversed(range(start_col, end_col)):
            result.append(array[end_row][col])
        
        for row in reversed(range(start_row+1, end_row)):
            result.append(array[row][start_col])


        start_row += 1
        end_row -= 1 
        start_col += 1
        end_col -= 1
    return result


spiral_traverse([[1,2,3,4],[12, 13, 14,5],[11,16,15,6],[10,9,8,7]])
```

```python
# Using recursion
# O(N)T and O(N)S : N is number of elements
def spiral_traverse(array):
    
    result = []
    start_col = 0
    end_col = len(array[0])-1
    start_row = 0
    end_row = len(array)-1

    spiral_fill(array, start_col, end_col, start_row, end_row,result)
    
    return result

def spiral_fill(array, start_col, end_col, start_row, end_row,result):
    if start_col > end_col and start_row > end_row:
        return
    for col in range(start_col, end_col+1):
        result.append(array[start_row][col])
    
    for row in range(start_row+1, end_row+1):
        result.append(array[row][end_col])

    for col in reversed(range(start_col, end_col)):
        result.append(array[end_row][col])
    
    for row in reversed(range(start_row+1, end_row)):
        result.append(array[row][start_col])


    start_row += 1
    end_row -= 1 
    start_col += 1
    end_col -= 1

    return spiral_fill(array, start_col, end_col, start_row, end_row,result)

spiral_traverse([[1,2,3,4],[12, 13, 14,5],[11,16,15,6],[10,9,8,7]])
```
