```python
# O(Nlog(N))T + O(Mlog(M))T : N is length of first array and M is length of second array
# O(1)S

def calc_smallest_difference(array_one, array_two):
    array_one.sort()
    array_two.sort()
    print(array_one, array_two)
    smallest_diff = float('inf')
    smallest_pair = []

    array_one_idx = 0
    array_two_idx = 0

    while array_one_idx < len(array_one) and array_two_idx < len(array_two):
        num_one = array_one[array_one_idx]
        num_two = array_two[array_two_idx]

        if num_one < num_two:
            current_diff = num_two - num_one
            array_one_idx += 1
        elif num_one > num_two:
            current_diff = num_one -num_two
            array_two_idx += 1
        else:
            return [num_one, num_two]

        if current_diff < smallest_diff:
            smallest_diff = current_diff
            smallest_pair = [num_one, num_two]
    return smallest_pair
        
array_one = [-1,5,10,20,28,3]
array_two = [26,134,135,15,17]
calc_smallest_difference(array_one, array_two)
```
