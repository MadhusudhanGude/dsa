```python
# O(Nlog(N))T + O(N**2)T ~= O(N**2)T : O(Nlog(N))T  to sort and O(N**2)T for the loops
# O(N)S 
def three_num_sum(array, target_sum):
    array.sort()    # O(Nlog(N))T 
    result = []

    for idx in range(len(array)-2):
        lowest_idx = idx+1
        highest_idx = len(array)-1

        while lowest_idx < highest_idx:
            current_sum = array[idx] + array[lowest_idx] + array[highest_idx]
            if current_sum == target_sum:
                result.append([array[idx] , array[lowest_idx] , array[highest_idx]])
                lowest_idx += 1
                highest_idx -= 1
            elif current_sum>target_sum:
                highest_idx -= 1
            elif current_sum<target_sum:
                lowest_idx += 1

    return result

array = [12,3,1,2,-6,5,-8,6]
target_sum = 0
three_num_sum(array, target_sum)
```
