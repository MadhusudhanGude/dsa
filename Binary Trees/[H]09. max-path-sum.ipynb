```python
# O(n)T and O(log(n))S
def max_path_sum(tree):
    _, max_path_sum = max_path_sum_helper(tree, parent=None)
    return max_path_sum


def max_path_sum_helper(tree):
    if tree is None:
        return 0,0
    
    value = tree.value

    left_max_as_branch, left_max_path_sum = max_path_sum_helper(tree.left)
    right_max_as_branch, right_max_path_sum = max_path_sum_helper(tree.right)

    max_child_sum_as_branch = max(right_max_as_branch, left_max_as_branch)
    max_sum_as_branch = max(value, max_child_sum_as_branch+value)
    
    max_path_as_root_node = (left_max_as_branch + value + right_max_as_branch, max_sum_as_branch)
    max_path_sum = max(left_max_path_sum, right_max_path_sum, max_path_as_root_node)

    return max_sum_as_branch, max_path_sum

```
