```python
# O(n)T and O(h)S 
def node_depths(node, depth=0):
    if node is None:
        return 0
    return depth + node_depths(node.left, depth+1) + node_depths(node.right, depth+1)
```

```python
# O(n)T and O(h)S 
def node_depths(root):
    stack = [{'node': root, 'depth':0}]
    result = 0
    while len(stack) > 0:
        node_info = stack.pop()
        node, depth = node_info['node'],  node_info['depth']
        if node is None:
            continue
        result += depth
        stack.append({'node': node.left, 'depth':depth+1})
        stack.append({'node': node.right, 'depth':depth+1})
    return result
```
