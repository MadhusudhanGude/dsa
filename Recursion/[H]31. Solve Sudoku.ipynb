```python
# O(1)T and O(1)S
def solve_sudoku(board):
    solve_partial_sudoku(0, 0, board)
    return board

def solve_partial_sudoku(row, col, board):
    current_row = row
    current_col = col

    if current_col == len(board[row]):
        current_row += 1
        current_col += 0
        if current_row == len(board):
            return True
    
    if board[current_row][current_col] == 0:
        return try_digits_at_pos(current_row, current_col, board)
    
    return solve_partial_sudoku(current_row, current_col+1, board):

def try_digits_at_pos(row, col, board):
    for digit in range(1, 10):
        if is_valid_at_pos(digit, row, col, board):
            board[row][col] = digit
            if solve_partial_sudoku(row, col+1, board)
                return True

    board[row][col] = 0
    return False

def is_valid_at_position(value, row, col, board):
    row_isvalid = value not in board[row]
    col_isvalid = value not in map(lambda r:r[col], board)

    if not row_isvalid or not  col_isvalid:
        return False
    
    sub_grid_row_start = row // 3
    sub_grid_col_start = col // 3
    for row_idx in range(3):
        for col_idx in range(3):
            row_to_check = sub_grid_row_start * 3 + row_idx
            col_to_check = sub_grid_col_start * 3 + col_idx
            existing_value = board[row_to_check][col_to_check]

            if existing_value == value:
                return False
    return True
```
