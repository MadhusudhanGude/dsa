```python
DIGIT_LETTERS = {
    "0": ["0"],
    "1": ["1"],
    "2": ["a", "b", "c"],
    "3": ["d", "e", "f"],
    "4": ["g", "h", "i"],
    "5": ["j", "k", "l"],
    "6": ["m", "n", "o"],
    "7": ["p", "q", "r", "s"],
    "8": ["t", "u", "v"],
    "9": ["w", "x", "y", "z"],
}


def mnemonic(phone_number):
    mnemonics = []
    current_mnemonic = ['0']*len(phone_number)
    mnemonic_helper(0, phone_number, current_mnemonic, mnemonics)
    return mnemonics

def mnemonic_helper(idx, phone_number, current_mnemonic, mnemonics):
    if idx == len(phone_number):
        mnemonics.append(''.join(current_mnemonic))
    else:
        for i in DIGIT_LETTERS[phone_number[idx]]:
            current_mnemonic[idx] = i
            mnemonic_helper(idx+1, phone_number, current_mnemonic, mnemonics)

mnemonic("1905")

```
