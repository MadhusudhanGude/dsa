```python
# O(n^2*n!)T and O(n*n!)S
def get_permutations(array):
    permutations = []
    helper(array, [], permutations)
    return permutations

def helper(array, current_permutation, permutations):
    if not len(array) and len(current_permutation):
        permutations.append(current_permutation)
    else:
        for i in range(len(array)):
            new_array = array[:i] + array[i+1:]
            new_permutation = current_permutation + [array[i]]
            helper(new_array, new_permutation, permutations)

get_permutations([1,2,3])
```

```python
# O(n*n!)T and O(n*n!)S
def get_permutations(array):
    permutations = []
    helper(0, array, permutations)
    return permutations

def helper(idx, array, permutations):
    if idx == len(array) - 1:
        permutations.append(array[:])
    else:
        for j in range(idx, len(array)):
            print(idx, j)
            swap(array, idx, j)
            helper(idx+1, array, permutations)
            swap(array, j, idx)

def swap(array, i, j):
    array[i], array[j] = array[j], array[i]

get_permutations([1,2,3])
```
