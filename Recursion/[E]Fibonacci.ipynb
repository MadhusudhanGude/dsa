```python
# O(2^N)T
# O(N)S
def fibonacci(index):
    if index == 0:
        return 0
    elif index == 1:
        return 1
    else:
        return fibonacci(index-1) + fibonacci(index-2)
fibonacci(24)
```

```python
# O(N)T
# O(N)S
def fibonacci(index, memoize={1 :0, 2:1}):
    if index in memoize:
        return  memoize[index]
    else:
        memoize[index] = fibonacci(index-1, memoize) + fibonacci(index-2, memoize)
        return memoize[index]
fibonacci(24)
```

```python
# O(1)S
# O(N)T
def fibonacci(index):
    last_two = [0,1]
    counter = 2
    while counter < index:
        next_fib = last_two[0]+last_two[1]
        last_two[0] = last_two[1] 
        last_two[1] = next_fib
        counter +=1
    return next_fib
fibonacci(4)
```
