```python
# O(n*k)T and O(n)S
# k - each recursion doing k elements calculation 
# n - no. of nodes
# RECURSION
def stair_case_traversal(height, max_steps):
    return num_of_ways_to_top(height, max_steps)

def num_of_ways_to_top(height, max_steps, memoize={0:1, 1:1}):

    if height in memoize:
        return memoize[height]
        
    num_of_ways = 0     
    for step in range(1, min(height, max_steps)+1):
        num_of_ways += num_of_ways_to_top(height-step, max_steps, memoize)

    memoize[height] = num_of_ways

    return num_of_ways

stair_case_traversal(4,2)
```

```python
# ITERATION: To avoid recursive solution
# O(n*k)T and O(n)S
def stair_case_traversal(height, max_steps):
    ways_to_top = [0] * (height + 1)
    ways_to_top[0] = 1
    ways_to_top[1] = 1

    for cur_height in range(2, height+1):
        step = 1
        while step <= max_steps and step <= cur_height:
            ways_to_top[cur_height] += ways_to_top[cur_height-step] 
            step += 1

    return ways_to_top[height]

stair_case_traversal(4,2)
```
