```python
# O(n*2^n)T and O(n*2^n)S
def power_set(array):
    subsets = [[]]
    for ele in array:
        for i in range(len(subsets)):
            subsets.append(subsets[i]+[ele])
    return subsets

power_set([1,2,3])
```

```python
# O(n*2^n)T and O(n*2^n)S
def power_set(array, idx = None):
    if idx is None:
        idx = len(array)-1
    elif idx < 0:
        return [[]]

    subsets = power_set(array, idx = idx-1)
    for i in range(len(subsets)):
        cur_subset = subsets[i]
        subsets.append(cur_subset + [array[idx]])
        
    return subsets

power_set([1,2,3])
```
