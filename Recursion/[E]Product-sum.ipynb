```python
# O(N)T : N is number of elements + number of elements in each matrix
# O(D)S: D is maximum depth of the inner array(3 in our example), as each inner array calc would get stored in stack trace
def product_sum(array, multiplier=1):
    sum = 0
    for ele in array:
        if type(ele) is list:
            sum += product_sum(ele, multiplier+1)
        else:
            sum += ele
    return sum*multiplier
```

```python
product_sum([5,2,[7,-1],3,[6,[-13,8],4]])
```
