```python
# O(N)S and O(N)T
def kth_ele(tree, k):
    array = []
    inorder_traverse(tree, array)
    return array[-k]

def inorder_traverse(tree, array):
    if tree is None:
        return

    inorder_traverse(tree.left, array)
    array.append(tree.value)
    inorder_traverse(tree.right, array)
```

```python
class TreeNodesVisited:
    def __init__(self, nodes_visited, latest_value):
        self.nodes_visited = nodes_visited
        self.latest_value = latest_value

def find_kthnode(tree, k):
    node_track = TreeNodesVisited(0, None)
    reverse_inorder_traverse(tree, k,  node_track)
    return node_track.latest_value

def reverse_inorder_traverse(tree, k, node_track):

    if tree is None or node_track.nodes_visited >=k:
        return

    reverse_inorder_traverse(tree.right, k, node_track)
    
    if node_track.nodes_visited < k :
        node_track.nodes_visited += 1
        node_track.latest_value += tree.value
        reverse_inorder_traverse(tree.left, k, node_track)
```
