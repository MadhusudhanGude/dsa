```python
class BST:

    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

    # O(n^2)T and O(h)S
    def reconstruct_bst(preorder_traversal_values):
        if len(preorder_traversal_values)==0:
            return None
        
        current_node_value = preorder_traversal_values[0]
        right_node_idx = len(preorder_traversal_values)

        for idx in range(1, len(preorder_traversal_values)): # change in value to handle elements with same value
            value = preorder_traversal_values[idx]
            if value >= current_node_value: # chnage in value to handle elements with same value
                right_node_idx = idx
                break
                
        right_tree = reconstruct_bst(preorder_traversal_values[right_node_idx:])
        left_tree = reconstruct_bst(preorder_traversal_values[1:right_node_idx])
        return BST(current_node_value, left_tree, right_tree)
```

```python
class TreeInfo:
    def __init__(self, value):
        self.root_idx = value

class BST:

    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

    # O(n)T and O(h)S
    def reconstruct_bst(preorder_traversal_values):
        tree = TreeInfo(0)
        return constructbst_from_range(float('inf'), float('-inf'), preorder_traversal_values, tree)


def  constructbst_from_range(lower_bound, upper_bound, preorder_traversal_values, tree):
    if tree.root_idx == len(preorder_traversal_values):
        return None
    
    root_value = preorder_traversal_values[tree.root_idx]
    if root_value >= upper_bound or root_value < lower_bound:
        return None
    
    tree.root_idx += 1

    left_subtree = constructbst_from_range(lower_bound, root_value, preorder_traversal_values, tree)
    right_subtree = constructbst_from_range(root_value, upper_bound, preorder_traversal_values, tree)

    return BST(root_value, left_subtree, right_subtree)
```
