```python
def selection_sort(array):
    current_index = 0
    while current_index < len(array) - 1:
        smallest_idx = current_index 
        for i in range(current_index+1, len(array)):
            if array[i] < array[current_index]:
                smallest_idx = i
        array[current_index], array[smallest_idx] = array[smallest_idx], array[current_index]
        current_index += 1

    return array

selection_sort([-34,4,43,89,9,0])
```
