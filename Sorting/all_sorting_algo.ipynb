```python
 def bubble_sort(array):
    for i in reversed(range(len(array))):
        no_swap = True
        
        for j in range(i):
            if array[j]>array[j+1]:
                array[j+1], array[j] = array[j], array[j+1]
                no_swap = False
        
        if no_swap:
            break

    return array

bubble_sort([32,23,1,9,21,-7, -1])
```

```python
# Select a number starting from 1st index and insert at correct position on left side
def insertion_sort(array):

    for i in range(1, len(array)):
        j = i
        while j>0 and array[j]<array[j-1]:
            array[j], array[j-1] = array[j-1], array[j]
            j -=1 
    return array

insertion_sort([-34,4,43,89,9,0])
```

```python
# Select a position and replace it with smallest number available from right side
def selection_sort(array):
    current_index = 0
    while current_index < len(array) - 1:
        smallest_idx = current_index 
        for i in range(current_index+1, len(array)):
            if array[i] < array[current_index]:
                smallest_idx = i
        array[current_index], array[smallest_idx] = array[smallest_idx], array[current_index]
        current_index += 1

    return array

selection_sort([-34,4,43,89,9,0])
```
