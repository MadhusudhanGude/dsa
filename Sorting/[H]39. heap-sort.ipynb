```python
# O(Nlog(N))T and O(1)S
def heap_sort(array):
    build_max_heap(array)
    for end_idx in reversed(range(1,len(array))):
        swap(array, 0, end_idx)
        sift_down(array, 0, end_idx-1)
    return array

def build_max_heap(array):
    first_parent_idx = (len(array) - 1) // 2
    for cur_idx in reversed(range(first_parent_idx+1)):
        sift_down(array, cur_idx, len(array) - 1 )

def sift_down(heap, cur_idx, end_idx):
    child_one_idx = cur_idx * 2 + 1
    while child_one_idx <= end_idx:
        child_two_idx = cur_idx * 2 + 2 if cur_idx * 2 + 2 <= end_idx else -1
        if child_two_idx > -1 and heap[child_two_idx] > heap[child_one_idx]:
            idx_to_swap = child_two_idx 
        else:
            idx_to_swap = child_one_idx
        if heap[idx_to_swap] > heap[cur_idx]:
            swap(heap, cur_idx, idx_to_swap)
            cur_idx = idx_to_swap
            child_one_idx = cur_idx * 2 + 1
        else:
            return 
                         
def swap(array, left, right):
    array[left], array[right] = array[right], array[left]


heap_sort([8,5,2,9,5,6,3])
```
