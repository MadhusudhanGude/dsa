```python
 def bubble_sort(array):
    for i in reversed(range(len(array))):
        no_swap = True
        
        for j in range(i):
            if array[j]>array[j+1]:
                array[j+1], array[j] = array[j], array[j+1]
                no_swap = False
        
        if no_swap:
            break

    return array

bubble_sort([32,23,1,9,21,-7, -1])
```
