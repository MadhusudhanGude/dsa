```python
# O(N)T and O(1)S
def find_largest_three(array):
    largest_num = [None, None, None]

    for ele in array:
        position = None
        if largest_num[2] is None or ele > largest_num[2]:
            position = 2
        elif largest_num[1] is None or ele > largest_num[1]:
            position = 1
        elif largest_num[1] is None or ele > largest_num[1]:
            position = 0

        if position is not None:
            for idx in range(position+1):
                if idx == position:
                    largest_num[idx] = ele
                else:
                    largest_num[idx] = largest_num[idx + 1]

    return largest_num

find_largest_three([141, 1, 17, -7, -17, -27, 18, 541, 8, 7, 7])
```
