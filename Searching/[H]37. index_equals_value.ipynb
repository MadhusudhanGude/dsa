```python
def index_equals_value(array):
    left_idx = 0
    right_idx = len(array)-1
    while left_idx <= right_idx:
        mid_idx = (left_idx + right_idx) // 2
        mid_value = array[mid_idx]
    
        # if mid_value > mid_idx:
        # IMPORTANT to have this condition instead of other way around to deal with increasing numbers
        if mid_value < mid_idx:
            # right_idx = mid_idx - 1
            left_idx = mid_idx + 1
        # First occurence of value = index
        elif mid_value == mid_idx and mid_idx == 0:
            return mid_idx
        # First occurence of value = index, by chekcing if the previous element is disorderd for sure
        elif mid_value == mid_idx and array[mid_idx - 1] < mid_idx - 1:
            return mid_idx
        else:
            # left_idx = mid_idx + 1
            right_idx = mid_idx - 1
    
    return -1

# index_equals_value([-5,-3,0,3,4,5,9])
index_equals_value([0,1,2,3,4,5])
```
