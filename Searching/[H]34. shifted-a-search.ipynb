```python
# O(1)S and O(log(n))T
def shifted_binary_Search(array, target):
    left = 0
    right = len(array)-1

    while left <= right:
        middle = (right + left) // 2
        left_num = array[left]
        middle_num = array[middle]
        right_num = array[right]
        if target == middle_num:
            return middle
        elif left_num <= middle_num:
            if target > left_num and target <= middle_num:
                right = middle - 1
            else:
                left = middle + 1
        else:
            if target < right_num and target >= middle_num:
                left = middle + 1
            else:
                right = middle - 1
    return -1
shifted_binary_Search([45,61,7172,73,0,1,21,33,45], 33)
```
