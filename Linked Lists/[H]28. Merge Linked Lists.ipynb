```python
# O(n+m)T and O(n+m)S
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next


def print_list(node):
    while node is not None:
        print(node.value, end=' ')
        node = node.next

def merge_linkedlist(head1, head2):
    dummy = Node(None)
    tail = dummy
    while head1 is not None and head2 is not None:
        if head1.value < head2.value:
            tail.next = head1
            head1 = head1.next
        else:
            tail.next = head2
            head2 = head2.next
        tail = tail.next
    if head1:
        tail.next = head1

    if head2:
        tail.next = head2
    
    return  dummy.next

    
node1 = Node(2,Node(6,Node(7,Node(8))))
node2 = Node(1,Node(3,Node(4,Node(5,Node(9,Node(10))))))
result = merge_linkedlist(node1, node2)
print_list(result)
print()
```

```python
# O(n+m)T and O(1)S
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next


def print_list(node):
    while node is not None:
        print(node.value, end=' ')
        node = node.next


def merge_linkedlist(head1, head2):

    lower = head1
    higher = head2

    # When the lower node is empty, just return highest
    if higher.value is None:
        return lower

    # When the higher node is empty, just return lower
    if lower.value is None:
        return higher

    if lower.value > higher.value:
        lower, higher = higher, lower

    # Start with lower
    result = lower

    while lower is not None and higher is not None:
        temp = lower
        while lower is not None and lower.value < higher.value:
            temp = lower
            lower = lower.next
        temp.next = higher
        lower, higher = higher, lower
    return result

node1 = Node(2,Node(6,Node(7,Node(8))))
node2 = Node(1,Node(3,Node(4,Node(5,Node(9,Node(10))))))
result = merge_linkedlist(node1, node2)
print_list(result)
print()

node1 = Node(1,Node(3,Node(4,Node(5,Node(9,Node(10))))))
node2 = Node(2,Node(6,Node(7,Node(8))))
result = merge_linkedlist(node1, node2)
print_list(result)
print()

node1 = Node(5,Node(7,Node(9)))
node2 = Node(3,Node(4,Node(8,Node(10))))
result = merge_linkedlist(node1, node2)
print_list(result)


# # https://www.youtube.com/watch?v=Xb4slcp1U38
# Merge Two Sorted Lists | Microsoft | Yahoo | Amazon
# 77,639 views  Oct 11, 2020  Check our Website: https://www.takeuforward.org/
```
