```python
class Node:
    def __init__(self, value, next=None):
        self.value= value
        self.next = next

def sum_f_ll(ll_one, ll_two):
    new_ll_head_pointer = Node(0)
    current_node = new_ll_head_pointer
    carry = 0

    node_one = ll_one
    node_two = ll_two

    while node_one is not None or node_two is not None or carry != 0:
        # print(node_one.value, node_two.value, carry)
        node_one_value = node_one.value if node_one is not None else 0
        node_two_value = node_two.value if node_two is not None else 0
        cuurent_sum = node_one_value + node_two_value + carry

        new_val = cuurent_sum % 10
        new_node = Node(new_val)
        current_node.next = new_node
        current_node = new_node

        carry = cuurent_sum // 10
        node_one = node_one.next if node_one is not None else None
        node_two = node_two.next if node_two is not None else None

    return new_ll_head_pointer.next


def printList(node):
    while(node != None):
        print(node.value, end=" ")
        node = node.next

ll_one = Node(2,Node(4,Node(7,Node(1))))
printList(ll_one)
print()
ll_two = Node(9,Node(4,Node(5)))
printList(ll_two)
print()

result = sum_f_ll(ll_one, ll_two)
printList(result)
print()

```
