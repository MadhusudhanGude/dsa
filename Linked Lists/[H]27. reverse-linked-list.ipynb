```python
# O(n)T and O(1)S
def reversed_ll(head):
    cur = head;
    prev = None

    while cur != None:
        temp = cur.next;
        cur.next = prev
        prev = cur
        cur = temp

    return prev


# https://www.youtube.com/watch?v=ugQ2DVJJroc
#anujbhaiya #dsaone
# Reverse a Linked List | Iteratively & Recursively | DSA-One Course #37
```
