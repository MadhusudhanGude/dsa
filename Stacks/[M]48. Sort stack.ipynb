```python
def sort_stack(stack):
    if len(stack) == 0:
        return 
    
    top = stack.pop()

    sort_stack(stack)

    insert_in_sorted_order(stack, top)

    return stack

def insert_in_sorted_order(stack, value):
    if len(stack) == 0 or stack[len(stack)-1] <= value:
        stack.append(value)
        return

    top = stack.pop()

    insert_in_sorted_order(stack, value)
    
    stack.append(top)
    return




sort_stack([0,34,-43,3,-8,10])
```
