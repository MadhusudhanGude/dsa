```python
# USING STACKS
def sunset_views(building, direction):
    step = 1 if direction == 'EAST' else -1
    current_idx = 0 if direction == 'EAST' else len(building) -1

    result_buildings = []
    while current_idx >= 0 and current_idx < len(building):
        while len(result_buildings) > 0 and building[result_buildings[-1]] <= building[current_idx]:
            result_buildings.pop()

        result_buildings.append(current_idx)
        current_idx += step

    if direction == 'WEST':
        return result_buildings[::-1]
    return result_buildings

print('Buildings at following indexes can see the SUN')
print(sunset_views([3,5,4,4,3,1,3,2], 'EAST'))
print(sunset_views([3,5,4,4,3,1,3,2], 'WEST'))
```
