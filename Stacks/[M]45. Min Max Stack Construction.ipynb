```python
class MinMaxStack:
    def __init__(self):
        self.min_max_stack = []
        self.stack = []

    def peek(self):
        return self.stack(len(self.stack)-1)

    def pop(self):
        if self.stack:
            self.stack.pop()
            self.min_max_stack.pop()

    def push(self, value):
        min_max = {'min': value, 'max': value}
        if self.min_max_stack:
            min_max['min'] = min(self.min_max_stack[-1]['min'],value)
            min_max['max']  = max(self.min_max_stack[-1]['max'],value)
        self.stack.append(value)
        self.min_max_stack.append(min_max)

    def get_min(self):
        return self.min_max_stack[-1]['min']

    def get_max(self):
        return self.min_max_stack[-1]['max']

    def print(self):
        for ele in self.stack:
            print(ele, end=' ')

stack = MinMaxStack()
stack.push(3)
stack.push(2)
stack.push(312)
stack.push(53)
stack.push(9)
stack.print()
print()
print(stack.get_min())
print(stack.get_max())
```
