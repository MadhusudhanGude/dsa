```python
# O(N)T and O(N)S

def shorten_path(path):
    starts_with_slash = path[0] == '/'
    tokens = filter(is_imp_token, path.split('/'))
    stack = []
    if starts_with_slash:
        stack.append('')
    # print(tokens)
    # print(stack)
    for token in tokens:
        # print(stack)
        if token == '..':
            # beginning of relative path or last one is double which can happen only incase of relative path
            if len(stack) == 0 or stack[-1] == '..':
                stack.append(token)
            # Not beginning of relative path
            elif stack[-1] != '':
                stack.pop()
        else:
            stack.append(token)
    if len(stack) == 1 and stack[0]== '':
        return '/'
    return '/'.join(stack)

def is_imp_token(token):
    return len(token)>0 and token != '.'

shorten_path('/foo/../test/../test/../foo//bar/./baz')
```
