```python
def balanced_brackets(string):
    opening_brackets = "({["
    closing_brackets = "]})"
    bracket_mapping  = {
        "]": "[",
        "}": "{",
        ")": "("
    }
    stack = []

    for char in string:
        if char in opening_brackets:
            stack.append(char)
        elif char in closing_brackets:
            if len(stack) > 0:
                last_item = stack.pop()
                if not last_item == bracket_mapping[char]:
                    return False
            else:
                return False

    return len(stack) == 0
    
balanced_brackets('(([]()()){})')
# balanced_brackets('(([]()()){)})')
```
