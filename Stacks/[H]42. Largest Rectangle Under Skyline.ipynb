```python
# O(N^2)T and O(1)S
def largest_rectange(buildings):
    max_area = 0 
    for idx in range(len(buildings)):
        current_height = buildings[idx]

        farthest_left = idx
        while farthest_left > 0 and buildings[farthest_left-1] >= current_height:
            farthest_left -= 1
        
        farthest_right = idx
        while farthest_right < len(buildings)-1 and buildings[farthest_right+1] >= current_height:
            farthest_right += 1

        current_area = (farthest_right-farthest_left+1) * current_height
        max_area = max(max_area, current_area)
    return max_area

largest_rectange([1,3,3,2,4,1,5,3,2])
        
```
