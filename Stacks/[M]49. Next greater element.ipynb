```python
    # O(n)T and O(n)S
def find_next_greater_ele(array):
    result = len(array) * [-1]
    stack = []
    for idx in range(len(array)*2):
        circular_idx = idx % len(array)
        while len(stack) > 0 and array[circular_idx] > array[stack[len(stack)-1]]:
            top_idx = stack.pop()
            result[top_idx] = array[circular_idx]
        
        stack.append(circular_idx)

    return result

find_next_greater_ele([2,5,-3,-4,6,7,2])
```
