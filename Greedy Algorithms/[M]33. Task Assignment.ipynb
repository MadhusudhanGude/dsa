```python
def task_assignments(k, tasks):
    paired_tasks = []
    sorted_tasks = sorted(tasks)
    task_duration_to_indices = get_task_duration_to_indices(tasks)

    for idx in range(k):
        least_duration_task = sorted_tasks[idx]
        least_duration_task_index = task_duration_to_indices[least_duration_task].pop()

        highest_duration_task = sorted_tasks[len(tasks) - 1 - idx]
        highest_duration_task_index = task_duration_to_indices[highest_duration_task].pop()

        paired_tasks.append([least_duration_task_index, highest_duration_task_index])
    return paired_tasks


def get_task_duration_to_indices(tasks):
    task_duration_to_indices = {}
    for idx, duration in enumerate(tasks):
        if duration in task_duration_to_indices:
            task_duration_to_indices[duration].append(idx)
        else:
            task_duration_to_indices[duration] = [idx]
    return task_duration_to_indices

task_assignments(3, [1,3,5,3,1,4])
```
