```python
# O(n^2)T and O(1)S
def valid_start_city(distances, fuel, mpg):
    num_of_cities = len(distances)

    for start_city_idx in range(num_of_cities):
        miles_remaining = 0

        for current_city_idx in range(start_city_idx, start_city_idx+num_of_cities):
            if miles_remaining < 0:
                continue
            
            current_city_idx = current_city_idx % num_of_cities
            fuel_from_current_city = fuel[current_city_idx]
            distance_to_next_city = distances[current_city_idx]
            miles_remaining += fuel_from_current_city * mpg - distance_to_next_city

        if miles_remaining >= 0:
            return start_city_idx
    
    return -1

valid_start_city([5,25,15,10,15], [1,2,1,0,3], 10)
```

```python
# O(n)T and O(1)S
def valid_start_city(distances, fuel, mpg):
    num_of_cities = len(distances)
    miles_remaining = 0

    index_of_starting_city = 0
    miles_remaining_at_starting_city = 0

    for city_idx in range(1, num_of_cities):
        distance_from_previous_city = distances[city_idx - 1]
        fuel_from_previous_city = fuel[city_idx - 1]
        miles_remaining += fuel_from_previous_city * mpg - distance_from_previous_city

        if miles_remaining < miles_remaining_at_starting_city:
            index_of_starting_city = city_idx
            miles_remaining_at_starting_city = miles_remaining
            
    return index_of_starting_city

valid_start_city([5,25,15,10,15], [1,2,1,0,3], 10)
```
