```python
# O(nlog(n))T and O(1)S
def class_photos(red_shirts, blue_shirts):
    red_shirts.sort(reverse=True)
    blue_shirts.sort(reverse=True)

    first_row = 'BLUE' if blue_shirts[0] < red_shirts[0] else 'RED'

    for idx in range(len(blue_shirts)):
        if first_row == 'RED':
            if red_shirts[idx] >= blue_shirts[idx]:
                return False
        else:
            if blue_shirts[idx] >= red_shirts[idx]:
                return False
    return True

class_photos([5,8,1,3,4], [6,9,2,4,5])
```
