```python
# O(nlog(n))T and O(1)S
def minimum_waiting_time(array):
    array.sort()
    waiting_time = 0
    total_waiting_time = 0

    for idx in range(1, len(array)):
        waiting_time += array[idx-1]
        total_waiting_time += waiting_time
    return total_waiting_time

minimum_waiting_time([3,2,1,2,6])
```

```python
# O(nlog(n))T and O(1)S
def minimum_waiting_time(array):
    array.sort()
    total_waiting_time = 0

    for idx, duration in enumerate(array):
        queries_left = len(array) - (idx + 1)
        total_waiting_time += (queries_left  * duration)
    return total_waiting_time

minimum_waiting_time([3,2,1,2,6])
```
